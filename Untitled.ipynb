### بعد از نرمال کردن دیتا کلاس 0 و 1 را جدا میکنیم و هم چنین به نسبت 80 به 20 دیتا تست و ترین را از همدیگر جدا میکنیم 

```python
data_normal_x = dataframe[['feature1', 'feature2']] = (dataframe[['feature1', 'feature2']] - dataframe[['feature1', 'feature2']].mean()) / dataframe[['feature1', 'feature2']].std()
data_normal_y = dataframe['label']
data_normal = pd.concat([data_normal_x, data_normal_y] , axis = 1)
data_normal_0_train = data_normal[data_normal['label'] == 0].values[:40,:]
data_normal_1_train = data_normal[data_normal['label'] == 1].values[:40,:]
data_normal_0_test = data_normal[data_normal['label'] == 0].values[40:,:]
data_normal_1_test = data_normal[data_normal['label'] == 1].values[40:,:]
data_normal_train = np.concatenate((data_normal_0_train ,data_normal_1_train ), axis = 0)
data_normal_test = np.concatenate((data_normal_0_test ,data_normal_1_test ), axis = 0)
data_normal_test_y = data_normal_test[:,2]
data_normal_test = np.concatenate((data_normal_test[:,:2] , np.ones(20).reshape(20,1)),axis=1)
data_normal_test = np.concatenate( ( data_normal_test[:10,:] , -data_normal_test[10:,:]) ,axis = 0 )
plt.scatter(data_normal_train [:40,0], data_normal_train [:40,1] , color = 'g')
plt.scatter(data_normal_train [40:,0], data_normal_train [40:,1] , color = 'r')
plt.suptitle('Train data class_0 & class_1')
plt.xlabel('Feature 1 ')
plt.ylabel('Feature 2')
```

## least square classification

```python
X_lsc = data_normal_train[:,:2]
y_lsc = data_normal_train[:,2].reshape(80,1)
# add bias to data 
X_lsc = np.concatenate((X_lsc,np.ones(80).reshape(80,1)),axis=1)
# class 1 should be negative 
X_lsc = np.concatenate((X_lsc[:40,:], -X_lsc[40:,:]), axis = 0)
# for every kind of data label should be +1 
y_lsc = np.ones(80).reshape(80,1)
w_lsc = np.linalg.inv(X_lsc.T @ X_lsc) @ X_lsc.T @ y_lsc
w_lsc
```

### line : w[0] * feature1 + w[1] * feature2 + w[2]

### خط پیدا شده را رسم میکنیم

```python
def calculate_line_least(X,W):
    return -(W[0]/W[1]) * X - (W[2]/W[1])
list_line_least_x = [-2,2]
list_line_least_y = [calculate_line_least(-2,w_lsc),calculate_line_least(2,w_lsc)]
plt.scatter(data_normal_train [:40,0], data_normal_train [:40,1] , color = 'g')
plt.scatter(data_normal_train [40:,0], data_normal_train [40:,1] , color = 'r')
plt.suptitle('Train data class_0 & class_1')
plt.xlabel('Feature 1 ')
plt.ylabel('Feature 2')
plt.plot(list_line_least_x,list_line_least_y,color='b')
```

# perceptron

```python
w0 = np.array([[0],[0],[1]])
epoc_perseptron = 1000
eta = 0.01
for i in range(0,epoc_perseptron):
    for j in range(0,80):
        if X_lsc[j,:] @ w0  >= 0 :
            continue
        else:
            w0 = w0 + eta * X_lsc[j,:].reshape(3,1)
w_perseptron = w0
w_perseptron
```

### خط پیدا شده را رسم میکنیم

```python
def calculate_line_perseptron(X,W):
    return -(W[0]/W[1]) * X - (W[2]/W[1])
list_line_perseptron_x = [-2,2]
list_line_perseptron_y = [calculate_line_perseptron(-2,w_perseptron),calculate_line_perseptron(2,w_perseptron)]
plt.scatter(data_normal_train [:40,0], data_normal_train [:40,1] , color = 'g')
plt.scatter(data_normal_train [40:,0], data_normal_train [40:,1] , color = 'r')
plt.suptitle('Train data class_0 & class_1')
plt.xlabel('Feature 1 ')
plt.ylabel('Feature 2')
plt.plot(list_line_perseptron_x,list_line_perseptron_y,color='b')
```

# Fisher

```python
def calculate_covariance(matrix):
    mean_values = np.mean(matrix, axis=0)
    mean_centered = matrix - mean_values
    covariance = np.dot(mean_centered.T, mean_centered)
    covariance /= matrix.shape[0]
    return covariance


def fisher_methos(A,B):
    meu1 = np.mean(A,axis=0)
    meu2 = np.mean(B,axis=0)
    s1 = (A.shape[0]-1) * calculate_covariance(A)
    s2 = (B.shape[0]-1) * calculate_covariance(B)
    sw = s1 + s2    
    sw_inv = np.linalg.inv(sw)
    mue3 = meu1 - meu2 
    sw_inv_1 = np.concatenate((sw_inv , np.ones(sw_inv.shape[1]).reshape(1 , sw_inv.shape[1])  ),axis=0)
    return sw_inv @ mue3.T

w_fisher11 = fisher_methos(X_lsc[:40,:2],-X_lsc[40:,:2])
def draw_fisher(w,bias):
    listt1 = [-2,2]
    listt2 = []
    listt2.append(-(w[0]/w[1]) * -2 + bias)
    listt2.append(-(w[0]/w[1]) * 2 + bias)
    plt.plot(listt1,listt2)

plt.scatter(X_lsc[:40,0],X_lsc[:40,1])
plt.scatter(X_lsc[40:,0],X_lsc[40:,1])
draw_fisher(w_fisher11,-0.80)
```

مقدار بایاس باید دستی تخمین شود و به همین دلیل مقادیر مختلف را امتحان میکنیم و بنظر میرسد که مقدار بایاس -0.8 مقدار مناسبی است

درباره روش پیاده سازی کردن فیشر بیشتر تحقیق کن چون جوابش با جوابی که از لیست اسکور بدست اوردیم دقیقا یکسان دراومده 

### Acuracy : Tp + Tn / (Tp + Tn + Fp + Fn)

```python
def acuracyeror(data,w): 
    count_negative = np.count_nonzero(data @ w < 0)
    return ((data @ w).shape[0] - count_negative) / (data @ w).shape[0]
acuracyeror(data_normal_test,w_perseptron)

x = np.array(["least square", "perseptron","Fisher"])
y = np.array([acuracyeror(data_normal_test,w_perseptron), acuracyeror(data_normal_test,w_lsc),1])
plt.ylabel('acuracy')
plt.bar(x,y)
plt.suptitle('Acuracy')
plt.show()
print('acuracy least square' , acuracyeror(data_normal_test,w_lsc) )
print('acuracy perseptron' , acuracyeror(data_normal_test,w_perseptron) )
print('acuracy Fisher' , 1 )
```
